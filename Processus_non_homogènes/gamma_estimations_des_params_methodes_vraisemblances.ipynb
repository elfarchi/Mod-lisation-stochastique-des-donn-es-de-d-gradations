```python
import numpy as  np
times = np.linspace(400,1000,2500)
```

```python
a  = 2
c = 2
b = 3

np.random.seed(42)
delta_x = np.array([np.random.gamma(a*(times[i]**c- times[i-1]**c), b, 1) for i in range(1,len(times)) ])




```

```python
from scipy.special import loggamma

def neg_log_vraisemblance(params, delta_x, t):
    a, b, c = params[0], params[1], params[2]
    
   
    
    # on aplatit delta_x
    dx = delta_x.flatten()
    

    # differences des t^c
    time_diff = t[1:len(t)]**c - t[0:len(t)-1]**c
    
    # les parametres de forme pour chaque observation
    shape = a * time_diff

    
    # log-vraisemblance
    lv = (shape - 1) * np.log(dx) - dx / b - shape * np.log(b) - loggamma(shape)

    
    return -np.sum(lv)


```

```python
from scipy.optimize import minimize

init_guess = [1, 1, 1]  


# optimisatiion
result = minimize(
    neg_log_vraisemblance,
    init_guess,
    args=(delta_x, times),
    method='L-BFGS-B'
)

# resultats
a_hat, b_hat, c_hat = result.x

print(f"a = {a_hat:.4f} (vrai valeur: {a})")
print(f"b = {b_hat:.4f} (vrai valeur: {b})")
print(f"c = {c_hat:.4f} (vrai valeur: {c})")
```
